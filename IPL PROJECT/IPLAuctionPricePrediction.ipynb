```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ipl_auction_df=pd.read_csv('IPLDATA.csv')
```

```python
print(ipl_auction_df)
```

```python
ipl_auction_df.info()
```

```python
ipl_auction_df['PLAYING ROLE'].unique()
```

```python
pd.get_dummies(ipl_auction_df['PLAYING ROLE'])
```

```python
X_features=[
    'AGE','PLAYING ROLE','T-RUNS','T-WKTS','ODI-RUNS-S','ODI-SR-B','ODI-WKTS','ODI-SR-BL','CAPTAINCY EXP','RUNS-S','HS','AVE','SR-B','SIXERS','RUNS-C','WKTS','AVE-BL','ECON','SR-BL']
```

```python
catagorical_features=['AGE','PLAYING ROLE','CAPTAINCY EXP']

```

```python
ipl_auction_encoded_df=pd.get_dummies(ipl_auction_df[X_features],columns=catagorical_features)
```

```python
ipl_auction_encoded_df
```

```python
ipl_auction_encoded_df.columns
```

```python
X=ipl_auction_encoded_df
```

```python
Y=ipl_auction_df['SOLD PRICE']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.7,random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
ypred=np.round(model.predict(x_test),2)
```

```python
ypred
```

```python
from sklearn import metrics
```

```python
np.round(metrics.r2_score(ypred,y_test),2)
```
